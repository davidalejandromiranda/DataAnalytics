## 1. Importar librerías

## 2. Planteamiento del problema

Realice una representación gráfica para:

+ **Interés compuesto:** La variación en el tiempo del interés pagado en un crédito a $n$ meses donde el capital inicial es $C$, el interés efectivo mensual $i$, el pago mensual es $M$ y está dado por un interés compuesto.

+ **Números aleatorios:** Valores aleatorios entre 0 y 1, suponiendo que cada valor tiene la misma probabilidad.

## 3. Cálculos y gráficas

### 3.1. Interés compuesto

El interés compuesto se obtiene cuando para un capital inicial $C$ se cobran intereses mensuales $i$ y se asume una cuota $M$ periódica constante que se paga durante $n$ periodos. Cada periodo parte de la cuota $C$ corresponde con los interés del periodo $I$ y el resto con pago a capital $M-I$.

La cuota mensual se calcula con la siguiente expresión:

$$M = C \frac{i(1+i)^n}{(1+i)^n-1}$$

```python
C = 25E6     # Monto inicial
i = 1.18/100 # Inerés mensual
n = 72       # Meses de plazo para pagar el crédito

M = C * i * ( 1 + i )**n / ( ( 1 + i )**n - 1 )

t = []
I = []
S = 1 * C # Saldo
for este_mes in range(n):
    interes_este_mes = S * i
    S = S - ( M - interes_este_mes )
    t.append(este_mes)
    I.append(interes_este_mes)

plt.figure(dpi=120)
plt.plot(t, I, 'ok')
plt.xlabel('Mes')
plt.ylabel('Interés por mes')

print('Cuota mensual:', np.round(M, 1))
print('Valor prestado:', C)
print('Valor total pgado:', np.round(n*M, 1))
print('Valor pagado en intereses:', np.round(sum(I), 1))
```

### 3.2. Números aleatorios

Para generar, por ejemplo, 500 números aleatorios entre 0 y 1, con igual probabilidad (distribución uniforme) se pueden obtener con la librería Numpy así:

$$numeros\_aleatorios = np.random.uniform(0, 1, size=500)$$

```python
numeros_aleatorios = np.random.uniform(0, 1, size=500)
plt.figure(dpi=120)
plt.plot(numeros_aleatorios, '.')
```

### 3.3. Diagrama de frecuencias

Una de las mejores maneras de obtener un diagrama de frecuencia es utilizando la librería **seaborn**, pero para ello es necesario colocar los datos en una tabla de **Pandas**. Las tablas de **Pandas** se conocen como *DataFrame* y se pueden construir de varias maneras, entre ellas usando diccionarios de **Python**.

```python
datos = pd.DataFrame({'Números aleatorios':numeros_aleatorios})
datos.tail
```

```python
sns.histplot(data=datos, bins=30)
```

```python
plt.figure(dpi=120)
_ = sns.histplot(data=datos, bins=30)
```

### 3.4. Distribución normal

La distribución normal está caracterizada por el valor medio y la desviación estándard. Por ejemplo, para obtener 10000 observaciones de una distribución  normal centrada en 0, con desviación estándard 1 se usa el siguiente comando en **Numpy**,

$$np.random.normal(loc=0, scale=1, size=10000)$$

```python
datos_normal = pd.DataFrame({'Datos distribución normal':
                            np.random.normal(loc=0, scale=1, size=10000)})
plt.figure(dpi=120)
_ = sns.histplot(data=datos_normal, bins=30)
```
